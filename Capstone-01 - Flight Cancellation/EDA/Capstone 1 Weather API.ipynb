## NOTE: API Calling Problems

When asked for the current weather, but the timestamp in caller forecast is for yesterday/tomorrow!
That means that Timezone is converted UNIX timestamps to local times incorrectly. For each UNIX timestamp, there is a corresponding local time for each and every different timezone. For example, the UNIX time 1466892000 corresponds to 6PM (18:00) in New York, 10PM (22:00) GMT, and midnight (00:00) of the following day in Amsterdam. When converting UNIX timestamps to local times, always use the timezone property of the API response, so that your software knows which timezone is the right one.

I have anticipated this problem and made the API call properly, I did sent the API call by taking the timezone differences in to account.
