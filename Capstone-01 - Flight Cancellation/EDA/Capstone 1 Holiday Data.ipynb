```python
import pandas as pd
from IPython.display import display
pd.options.display.max_columns = None
import datetime
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
flights=pd.read_csv('Capstone 1 - Preparing Data(W).csv',low_memory=False,dtype=str)
```

```python
flights.MONTH.unique()
```

```python
holiday=pd.read_csv('us-federal-holidays-2011-2020-Copy1.csv')
holiday.head()
```

```python
holiday_2015 = holiday[holiday.Date.str.contains('2015')]
holiday_2015.Date=['01/01/2015',
 '01/19/2015',
 '02/16/2015',
 '05/25/2015',
 '07/03/2015',
 '09/07/2015',
 '10/12/2015',
 '11/11/2015',
 '11/26/2015',
 '12/25/2015']
```

```python
holiday_dict = holiday_2015[['Holiday']].set_index(holiday_2015.Date).to_dict()
```

```python
holiday_dict=holiday_dict['Holiday']
```

```python
holiday_dict
```

```python
flights['IsHoliday']=flights.DATE
flights['IsHoliday']=flights['IsHoliday'].str.slice(5,7)+"/"+flights['IsHoliday'].str.slice(8,10)+"/"+flights['IsHoliday'].str.slice(0,4)
```

```python
days =list(holiday_dict.values())
```

```python
flights['IsHoliday']=flights['IsHoliday'].replace(holiday_dict)
```

```python
flights['IsHoliday'][flights['IsHoliday'].isin(days)].unique()
```

```python
flights.DATE[flights.DATE.str.contains('10-14')]
```

```python
flights['Holiday']=flights['IsHoliday']
```

```python
flights['IsHoliday'][flights['IsHoliday'].isin(days)]=1
flights['IsHoliday'][flights['IsHoliday']!=1]=0
```
