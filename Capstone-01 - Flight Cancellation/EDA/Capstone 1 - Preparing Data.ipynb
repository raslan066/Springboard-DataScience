### 2. Computation Time Adjustments 

I will get the most common airports in terms of air traffic for both Origin and Destination Airports. This will reduce the amount of computation. All the steps and procedures will be used on this 10 most common airports can be applied on the rest of the airports as well  Here is the procedure below.
- Get the size of the airport for both Origin and Destination airports. Airports that are the most busiest is also the busiest destination airports as well
- Filter the data set that only includes most common airports for both Origin and Destination airports

```python
flights.MONTH[flights.ORIGIN_AIRPORT=='LAX'].unique()
```

```python
#Selecting top 10 Origin Airport 
top_10_ORIGIN_AIRPORT= flights.ORIGIN_AIRPORT.groupby(flights.ORIGIN_AIRPORT).size().sort_values(ascending=False)[:10].to_dict()
top_10_ORIGIN_AIRPORT
```

#### 3.4 Correcting Scheduled Departure Dates and Times
- Simply combining date and times for the departure is not going to be the correct procedure here. Because due to delays the days of the flight changes. This is something need to be corrected.

#### Departure time

```python
# In order to work with the datetime object I need to convert it to datetime.
flights['SCHEDULED_DEPARTURE_DT']=pd.to_datetime(flights['SCHEDULED_DEPARTURE_DT'])
```

```python
# Had a temp column to keep time in epoch format this makes the computation a lot faster. 
import datetime, time
flights['temp'] = flights['SCHEDULED_DEPARTURE_DT'].apply(lambda x : time.mktime(x.timetuple())) 
```

```python
#Added the delays to Scheduled time and convert it back to proper Date format with accurate year day and time
flights['DEPARTURE_TIME_DT']= flights['temp']+flights['DEPARTURE_DELAY']*60
flights['DEPARTURE_TIME_DT']=flights['DEPARTURE_TIME_DT'][flights['DEPARTURE_TIME_DT'].notnull()].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
flights['DEPARTURE_TIME_DT'].head()
```


#### Wheel Off time

```python
flights['TAXI_OUT']=flights['TAXI_OUT'][flights['TAXI_OUT'].notnull()].astype(float).astype(int)
```

```python
flights['DEPARTURE_TIME_DT']=pd.to_datetime(flights['DEPARTURE_TIME_DT'])
flights['temp']=''
flights['temp'] = flights['DEPARTURE_TIME_DT'][flights['DEPARTURE_TIME_DT'].notnull()].apply(lambda x : time.mktime(x.timetuple()))
flights['WHEELS_OFF_DT']= flights['temp']+flights['TAXI_OUT']*60
flights['WHEELS_OFF_DT']=flights['WHEELS_OFF_DT'][flights['WHEELS_OFF_DT'].notnull()].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
flights['WHEELS_OFF_DT'].head()
```

#### 3.4.1 Correcting Scheduled Arrival Dates and Times

```python
flights.head()
```

```python
flights['AIR_TIME']=flights['AIR_TIME'][flights['AIR_TIME'].notnull()].astype(float).astype(int)
flights['SCHEDULED_TIME']=flights['SCHEDULED_TIME'][flights['SCHEDULED_TIME'].notnull()].astype(float).astype(int)
flights['ELAPSED_TIME']=flights['ELAPSED_TIME'][flights['ELAPSED_TIME'].notnull()].astype(float).astype(int)
flights['WHEELS_OFF_DT']=pd.to_datetime(flights['WHEELS_OFF_DT'])
```

```python
flights['temp']=''
flights['temp'] = flights['WHEELS_OFF_DT'][flights['WHEELS_OFF_DT'].notnull()].apply(lambda x : time.mktime(x.timetuple()))
flights['WHEELS_ON_DT']= flights['temp']+flights['AIR_TIME']*60
flights['WHEELS_ON_DT']=flights['WHEELS_ON_DT'][flights['WHEELS_ON_DT'].notnull()].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
flights['WHEELS_ON_DT'].head()
```

```python
flights['temp']=''
flights['temp'] = flights['WHEELS_ON_DT'][flights['WHEELS_ON_DT'].notnull()].apply(lambda x : time.mktime(x.timetuple()))
flights['ARRIVAL_TIME_DT']= flights['temp']+flights['TAXI_IN']*60
flights['ARRIVAL_TIME_DT']=flights['ARRIVAL_TIME_DT'][flights['ARRIVAL_TIME_DT'].notnull()].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d %H:%M:%S'))
flights['ARRIVAL_TIME_DT'].head()
```
